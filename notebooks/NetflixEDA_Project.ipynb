```python
df['duration_minutes'] = np.where(df['duration'].str.contains("min" , na = False) , df['duration_number'] , np.nan)
df['seasons'] = np.where(df['duration'].str.contains('Season' , na = False) , df['duration_number'] , np.nan)
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df['genre_list'] = df['listed_in'].str.split(",")
df['genre_list'] = df['genre_list'].apply(lambda x: [g.strip() for g in x])
```

```python
df['primary_genre'] = df['genre_list'].apply(lambda x : x[0] if len(x) > 0 else None)
```

```python
year_count = df['year_added'].value_counts().sort_index()
plt.figure(figsize=(8,5))
plt.plot(year_count.index , year_count.values , marker ='o')
plt.xlabel("Year Added")
plt.ylabel("Number of Titles Added")
plt.grid(alpha = 0.3)
plt.title("Number of Netflix titles added each year")
plt.show()
```

```python
genre_counts = df['primary_genre'].value_counts().head(10)
plt.figure(figsize = (10,5))
plt.bar(genre_counts.index , genre_counts.values)
plt.title("Top 10 Primary Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation = 45 , ha = 'right')
plt.grid(axis = 'y', alpha =0.3)
plt.show()
```

```python
type_counts = df['type'].value_counts()
plt.figure(figsize =(6,6))
plt.pie(type_counts.values , labels = type_counts.index , autopct ='%1.1f%%')
plt.title("Distribution of Movies vs TV Shows on Netflix")
plt.show()
```

```python
df['primary_country'] = df['country'].apply(lambda x : x.split(",")[0].strip())
country_counts = df['primary_country'].value_counts(dropna = False).head(20)
plt.figure(figsize = (20,10))
plt.bar(country_counts.index , country_counts.values )
plt.xlabel('Country')
plt.ylabel('Number of Title')
plt.title('Top 20 countries producing netflix content')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
rating_count = df['rating'].value_counts()
plt.figure(figsize =(10,5))
plt.bar(rating_count.index , rating_count.values )
plt.title("Distribution of Content Ratings on Netflix")
plt.xlabel('Rating Category')
plt.ylabel('Number of Titles')
plt.grid(alpha = 0.3)
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
movies = df[df['type'] == "Movie"]
tv_shows = df[df['type'] == "TV Show"]
movie_genres = movies['primary_genre'].value_counts().head(10)
tv_genres = tv_shows['primary_genre'].value_counts().head(10)
plt.figure(figsize=(14,6))

plt.subplot(1, 2, 1)
plt.bar(movie_genres.index, movie_genres.values)
plt.title("Top Genres in Movies")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.subplot(1, 2, 2)
plt.bar(tv_genres.index, tv_genres.values, color='orange')
plt.title("Top Genres in TV Shows")
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
df_cast = df[df['cast'].notnull()].copy()
df_cast['cast_list']= df_cast['cast'].str.split(",")
actors_exploded = df_cast.explode('cast_list')
actors_exploded['cast_list'] = actors_exploded['cast_list'].str.strip()
top_actors = actors_exploded['cast_list'].value_counts().head(20)
plt.figure(figsize = (12,6))
plt.barh(top_actors.index , top_actors.values)
plt.xlabel("Number of Titles")
plt.ylabel("Actor")
plt.title("Top 20 Most Featured Actors on Netflix")
plt.grid(axis='x', alpha=0.3)
plt.show()
```

```python
df_director = df[df['director'].notnull()].copy()
df_director['director_list'] = df_director['director'].str.split(",")
directors_exploded = df_director.explode('director_list')
directors_exploded['director_list'] = directors_exploded['director_list'].str.strip()
top_directors = directors_exploded['director_list'].value_counts().head(20)
plt.figure(figsize =(12,6))
plt.barh(top_directors.index , top_directors.values)
plt.xlabel('Number Of Titles')
plt.ylabel('Director')
plt.title('Top 20 directors on Netflix')
plt.grid(axis = 'x' , alpha = 0.3)
plt.show()
```
